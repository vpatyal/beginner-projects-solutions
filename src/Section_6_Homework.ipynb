```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
#plt.rcParams['figure.figsize'] = 8,4
import warnings
warnings.filterwarnings('ignore')
```

```python
df_movies = pd.read_csv("C:\\Users\\vpatyal\\Dropbox\\learn\\PythonPrograms\\jupyterNBcode\\data\\P4-Section6-Homework-Dataset.csv", encoding='latin1')
```

```python
df_movies.columns = ['DayofWeek', 'Director', 'Genre', 'MovieTitle', 'ReleaseDate',
       'Studio', 'AdjustedGrossMillions', 'BudgetMillions', 'GrossMillions',
       'IMDbRating', 'MovieLensRating', 'OverseasMillions', 'OverseasPercentage',
       'ProfitMillions', 'ProfitPercentage', 'Runtime', 'USMillions',
       'GrossUSPercentage']
```

```python
len(df_movies)
```

```python
df_movies['Genre'] = df_movies['Genre'].astype('category')
df_movies['Studio'] = df_movies['Studio'].astype('category')
```

```python
df_movies2 = df_movies[df_movies['Genre'].isin(['action','comedy','adventure','animation','drama'])]
```

```python
df_movies3 = df_movies2[df_movies2['Studio'].isin(['Fox','WB','Buena Vista Studios','Paramount Pictures','Sony', 'Universal'])].reset_index()
```

```python
(df_movies3.Studio.unique())
```

```python
df_movies3 = df_movies3.drop('index', axis=1)
```

```python
len(df_movies3)
```

```python
sns.set(style = 'darkgrid', palette='muted', color_codes=True)
```

```python
ax = sns.boxplot(data = df_movies3, x='Genre', y='GrossUSPercentage', orient='v', color='lightgrey', showfliers=False)

plt.setp(ax.artists, alpha=0.5)

sns.stripplot(data = df_movies3, x='Genre', y='GrossUSPercentage', jitter=True, size=6, linewidth=0, hue='Studio', alpha=0.7)

ax.axes.set_title('Domestic Gross % by Genre', fontsize=30)
ax.set_xlabel('Genre', fontsize=20)
ax.set_ylabel('Gross % US', fontsize=20)

ax.legend(bbox_to_anchor=(1.05,1),loc=2)
```
